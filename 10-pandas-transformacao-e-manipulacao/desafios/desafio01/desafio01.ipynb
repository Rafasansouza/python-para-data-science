```python
import pandas as pd
```

```python
df = pd.read_json('dados_vendas_clientes.json')
df
```

```python
df_normalizado = pd.json_normalize(df['dados_vendas'])
df_normalizado
```

```python
colunas = list(df_normalizado.columns)
colunas
```

```python
df_explode = df_normalizado.explode(colunas[1:])
df_explode
```

```python
df_explode.info()
```

```python
df_explode['Valor da compra'] = df_explode['Valor da compra'].apply(lambda x: x.replace('R$ ', "").replace(',', ".").strip())
df_explode
```

```python
import numpy as np
df_explode['Valor da compra'] = df_explode['Valor da compra'].astype(np.float64)
df_explode.info()
```
