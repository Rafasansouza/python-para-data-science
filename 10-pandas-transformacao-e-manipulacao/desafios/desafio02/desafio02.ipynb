```python
import pandas as pd
```

```python
df = pd.read_json('dados_locacao_imoveis.json')
df
```

```python
df_normalizado = pd.json_normalize(df['dados_locacao'])
df_normalizado
```

```python
colunas = list(df_normalizado.columns)
colunas
```

```python
df_explode = df_normalizado.explode(colunas[1:])
df_explode
```

```python
df_explode.info()
```

```python
df_explode['valor_aluguel'] = df_explode['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace(',', '.').replace(' reais', '').strip())
df_explode
```

```python
import numpy as np
df_explode['valor_aluguel'] = df_explode['valor_aluguel'].astype(np.float64)
df_explode.info()
```

```python
df_explode['apartamento'] = df_explode['apartamento'].str.replace(r'\(.*?\)', '', regex=True)
df_explode.head()
```
