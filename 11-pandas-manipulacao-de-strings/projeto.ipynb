```python
import pandas as pd
```

```python
df = pd.read_csv('banco_clientes.csv')
df
```

Padronizar letras

```python
df['nome'] = df['nome'].str.upper()
```

```python
df
```

```python
df.telefone.str.replace("(", "", regex=False).str.replace(")", "", regex=False)
```

```python
df['telefone'] = df.telefone.str.replace("(", "", regex=False).str.replace(")", "", regex=False)
df
```

```python
df.endereco.str.extract("(\d{5}-\d{3})")
```

```python
df['cep'] = df.endereco.str.extract("(\d{5}-\d{3})")
df
```

```python
df['endereco'] = df.endereco.str.replace("(\d{5}-\d{3})", "", regex=True)
df
```

```python
df.endereco.str.replace(",", "").str.strip().str.split('\s(?=\d+$)', expand=True)
```

```python
df[["rua", "numero"]] = df.endereco.str.replace(",", "").str.strip().str.split('\s(?=\d+$)', regex=True, expand=True)
df
```

```python
df = df.drop(['endereco'], axis=1)
df
```

```python
df['rua'] = df['rua'].str.upper()
df
```

```python
df.to_csv('banco_clientes_tratado.csv', index=False)
```
